```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt

# Generate data for two clusters


# Generate data for two blobs with separate density
cluster1_x, cluster1_y = make_blobs(n_samples=300, centers=[[1, 1]], cluster_std=0.5, )

cluster2_x = np.random.uniform(low=2.5, high=8.5, size=(100, 2))
cluster3_x = np.random.uniform(low=0, high=7, size=(20, 2))

# Combine the datasets
combined_x = np.concatenate((cluster1_x, cluster2_x))
combined_x = np.concatenate((combined_x, cluster3_x))

# Create the scatter plot
plt.scatter(combined_x[:, 0], combined_x[:, 1],)

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Scatter Plot with Separate Clusters')



# Show the plot
plt.show()

```

```python
from sklearn.neighbors import NearestNeighbors

# Create an instance of the NearestNeighbors class
knn = NearestNeighbors(n_neighbors=5)

# Fit the model to the data
knn.fit(combined_x)

# Find the distances and indices of the k nearest neighbors for each data point
distances, indices = knn.kneighbors(combined_x)

# Set the threshold for outlier detection
threshold = .55




# Identify the outliers based on the average distance to the k nearest neighbors
outlier_mask = distances.mean(axis=1) > threshold
combined_x_cleaned = combined_x[~outlier_mask]
# Plot the outliers as stars
plt.scatter(combined_x_cleaned[:, 0], combined_x_cleaned[:, 1], label='Inliers')

plt.scatter(combined_x[outlier_mask, 0], combined_x[outlier_mask, 1], marker='*', color='red', label='Outliers')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('k-Nearest Neighbors Outlier Detection Example')

# Add legend
plt.legend()

# Show the plot
plt.show()

# Plot the outliers as stars
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Generate some data
np.random.seed(42)
X = 0.3 * np.random.randn(100, 2)
X_train = np.r_[X + 2, X - 2]
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))

# Fit Isolation Forest
clf = IsolationForest(random_state=42)
clf.fit(X_train)

# Predict outliers
y_pred_train = clf.predict(X_train)
y_pred_outliers = clf.predict(X_outliers)

# Plot the results
plt.figure(figsize=(10, 6))

# Inliers
plt.scatter(X_train[:, 0], X_train[:, 1], c='white', edgecolor='k', label='Inliers')

# Outliers
plt.scatter(X_outliers[:, 0], X_outliers[:, 1], c='red', edgecolor='k', label='Outliers')

# Decision boundary
xx, yy = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='black')

plt.title('Isolation Forest')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Generate data in an elliptical shape
def generate_elliptical_data(num_samples, center, cov_matrix):
    return np.random.multivariate_normal(center, cov_matrix, num_samples)

# Define parameters for the elliptical shape
num_samples = 200
center = [0, 0]  # Center of the ellipse
cov_matrix = [[2, 1], [4, 3]]  # Covariance matrix

# Generate elliptical data
X = generate_elliptical_data(num_samples, center, cov_matrix)

# Compute mean and covariance matrix
mean = np.mean(X, axis=0)
cov_inv = np.linalg.inv(np.cov(X.T))

# Calculate Mahalanobis distance
def mahalanobis_distance(x, mean, cov_inv):
    diff = x - mean
    return np.sqrt(np.dot(np.dot(diff, cov_inv), diff.T))

mahalanobis_distances = [mahalanobis_distance(x, mean, cov_inv) for x in X]

# Calculate Euclidean distance
euclidean_distances = np.linalg.norm(X - mean, axis=1)

# Set the threshold for Mahalanobis distance
threshold = np.sqrt(chi2.ppf(0.975, df=2))  # 95% confidence interval for 2 dimensions

# Identify outliers based on Mahalanobis distance
mahalanobis_outliers = np.where(mahalanobis_distances > threshold)[0]

# Plot the results
plt.figure(figsize=(10, 6))

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c='blue', label='Data Points')

# Highlight outliers based on Mahalanobis distance
plt.scatter(X[mahalanobis_outliers, 0], X[mahalanobis_outliers, 1], c='red', label='Mahalanobis Outliers')

# Add legend and labels
plt.title('Elliptical Data with Mahalanobis Outliers')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Add ellipse representing Mahalanobis distance
theta = np.linspace(0, 2 * np.pi, 100)
x_ellipse = threshold * np.cos(theta) + mean[0]
y_ellipse = threshold * np.sin(theta) + mean[1]
plt.plot(x_ellipse, y_ellipse, color='green', linestyle='--', label='Mahalanobis Distance')

plt.axis('equal')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
from matplotlib.patches import Ellipse

# Generate data in an elliptical shape
def generate_elliptical_data(num_samples, center, cov_matrix):
    return np.random.multivariate_normal(center, cov_matrix, num_samples)

# Define parameters for the elliptical shape
num_samples = 200
center = [0, 0]  # Center of the ellipse
cov_matrix = [[4, 0.5], [0.5, .2]]  # Covariance matrix (thinner)

# Generate elliptical data
X = generate_elliptical_data(num_samples, center, cov_matrix)

# Compute mean and covariance matrix
mean = np.mean(X, axis=0)
cov_inv = np.linalg.inv(np.cov(X.T))

# Calculate Mahalanobis distance
def mahalanobis_distance(x, mean, cov_inv):
    diff = x - mean
    return np.sqrt(np.dot(np.dot(diff, cov_inv), diff.T))

mahalanobis_distances = [mahalanobis_distance(x, mean, cov_inv) for x in X]

# Calculate Euclidean distance
euclidean_distances = np.linalg.norm(X - mean, axis=1)

# Set the threshold for Mahalanobis distance
threshold_mahalanobis = np.sqrt(chi2.ppf(0.975, df=2))  # 95% confidence interval for 2 dimensions

# Identify outliers based on Mahalanobis distance
mahalanobis_outliers = np.where(mahalanobis_distances > threshold_mahalanobis)[0]

# Set the threshold for Euclidean distance
threshold_euclidean = np.percentile(euclidean_distances, 95)  # 95th percentile of Euclidean distances

# Identify outliers based on Euclidean distance
euclidean_outliers = np.where(euclidean_distances > threshold_euclidean)[0]

# Plot the results
plt.figure(figsize=(10, 6))

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c='blue', label='Data Points')

# Plot ellipse representing decision boundary for Mahalanobis distance
cov_eigenvalues, cov_eigenvectors = np.linalg.eig(cov_matrix)
angle = np.arctan2(cov_eigenvectors[1, 0], cov_eigenvectors[0, 0])
width, height = 2 * np.sqrt(chi2.ppf(0.975, df=2) * cov_eigenvalues)
ellipse = Ellipse(xy=mean, width=width, height=height, angle=np.degrees(angle), edgecolor='green', fill=False, linestyle='--', label='Mahalanobis Distance')

plt.gca().add_patch(ellipse)

# Highlight outliers based on Mahalanobis distance
plt.scatter(X[mahalanobis_outliers, 0], X[mahalanobis_outliers, 1], c='red', label='Mahalanobis Outliers')

# Highlight outliers based on Euclidean distance
plt.scatter(X[euclidean_outliers, 0], X[euclidean_outliers, 1], c='yellow', label='Euclidean Outliers')

# Add legend and labels
plt.title('Elliptical Data with Outliers and Decision Boundaries')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.axis('equal')
plt.legend()
plt.show()
```
